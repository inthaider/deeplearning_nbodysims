### TRAIN DATA

```python

sigma_batch = 300
sigma_max = 2.0
sigma_min = 0.5
sigma_step = (sigma_max - sigma_min) / (test_size/sigma_batch - 1)

sig_values = np.empty((2100))

sigma = sigma_min
for i in range(2100):
    sig_values[i] = sigma
    if (i+1) % sigma_batch == 0:
        sigma += sigma_step
#         print(j, ") sigma set to:                        sigma=", sigma)
        alpha = alpha_min
```

```python
train_size = 4800
N = 2000
nt = 500
n_bins = 100

sigma_batch = 1200
sigma_max = 2.0
sigma_min = 0.5
sigma_step = (sigma_max - sigma_min) / (train_size/sigma_batch - 1)

alpha_batch = 120
alpha_max = 0.65
alpha_min = 0.01
alpha_step = (alpha_max - alpha_min) / (sigma_batch/alpha_batch - 1)

amp_batch = 10
amp_max = 9.
amp_min = 0.
amp_step = (amp_max - amp_min) / (alpha_batch/amp_batch - 1)

x_min = -30.
x_max = 25.
x_range = x_max - x_min
v_min = -8.
v_max = 6.
v_range = v_max - v_min


# density array is created with dimensions:
# (input & output states (0 or 1), x or v (0 or 1), training data size, num. bins + pert param + alpha)
train_density = numpy.empty((2, 2, train_size, n_bins+2))
train_density[1, :, :, -2:] = 0
energy = numpy.empty((train_size, nt+1))

amp_count = 0
sigma = sigma_min
alpha = alpha_min # "live" fraction
amp = amp_min # pert amp

# print("INITIAL:                                 sigma=", sigma, ", alpha=", alpha, ", amp=", amp, "\n\n")
for j in trange(train_size, desc='Training data progress'):
    numpy.random.seed(j+100000)
    
#     print(j, ") First instance in loop:                sigma=", sigma, ", alpha=", alpha, ", amp=", amp)
    
    # (input & output states, phase space degrees of freedom (x & v), no. of slabs)
    train = numpy.empty((2, 2, N))
        
    if (j) % amp_batch == 0:
        amp = amp_step*amp_count
        amp_count +=1
#         print(j, ") amp set to:                              amp=", amp)
        hp = HernquistPotential(amp=amp, a=1.)
#         print(j, ") POTENTIAL CALCULATED USING:              amp=", amp)
#     else:
#         print(j, ") alpha stays:                           alpha=", alpha)
#         print(j, ") amp stays:                               amp=", amp)
#         print(j, ") PREVIOUS POTENTIAL USED")
    
    train_density[0, :, j, -1] = alpha
    train_density[0, :, j, -2] = amp
    
#     print(j, ") STORED:                                alpha=", train_density[0, :, j, -1], ", amp=", train_density[0, :, j, -2])


    def extF(x, t):
        z = abs(Jo.z(t)-x)
        return hp.zforce(0., z)

    

    # compute zh based on sigma and totmass
    totmass= 1. # Sigma above
    zh= sigma**2./totmass # twopiG = 1. in our units
    tdyn= zh/sigma
    x= numpy.arctanh(2.*numpy.random.uniform(size=N)-1)*zh*2.
#     print(j, ") VELOCITY DISTR. CALCULATED USING:     sigma=", sigma)
    v= numpy.random.normal(size=N)*sigma
    v-= numpy.mean(v) # stabilize
    m= numpy.ones_like(x)/N#*(1.+0.1*(2.*numpy.random.uniform(size=N)-1))
    
    
#     print(j, ") SELF-GRAVITATION CALCULATED USING:     alpha=", alpha)
    # Adjust masses to only represent alpha of the mass
    m*= alpha
    # 1-alpha in the mass is then given by the external force
    sigma2= sigma**2.
    def tot_force(x,t):
        return extF(x, t)-(1.-alpha)*sigma2*numpy.tanh(0.5*x/zh)/zh
    
    
    g= wendy.nbody(x, v, m, 0.01*tdyn, approx=True, nleap=10, t0=-2.*dt, ext_force=tot_force)
    
    
    train[0, 0, :]= x
    train[0, 1, :]= v
    energy[j, 0]= wendy.energy(x,v,m)
    
    for ii in arange(nt):
        tx,tv = next(g)
        energy[j, ii+1]= wendy.energy(tx,tv,m)
        
    train[1, 0, :]= tx
    train[1, 1, :]= tv
    
    
    
    initial_xdensity, bins = np.histogram(train[0, 0, :], bins = n_bins, range=[x_min, x_max])
    final_xdensity, bins = np.histogram(train[1, 0, :], bins = n_bins, range=[x_min, x_max])
    train_density[0, 0, j, :-2] =  initial_xdensity
    train_density[1, 0, j, :-2] =  final_xdensity

    initial_vdensity, bins = np.histogram(train[0, 1, :], bins = n_bins, range=[v_min, v_max])
    final_vdensity, bins = np.histogram(train[1, 1, :], bins = n_bins, range=[v_min, v_max])
    train_density[0, 1, j, :-2] =  initial_vdensity
    train_density[1, 1, j, :-2] =  final_vdensity
    
    if (j+1) % alpha_batch == 0:
        alpha += alpha_step
#         print(j, ") alpha set to:                        alpha=", alpha)
        amp_count = 0.
        
    if (j+1) % sigma_batch == 0:
        sigma += sigma_step
#         print(j, ") sigma set to:                        sigma=", sigma)
        alpha = alpha_min
#         print(j, ") alpha set to 0:                        alpha=", alpha)
            
#     print("\n\n")
```

## Density Profile Datasets with Fixed Perturbations and Fully Self-Gravitating Disc

```python
train_size = 2000
N = 1001
nt = 500
n_bins = 50
x_min = -15.
x_max = 15.
x_range = x_max - x_min
v_min = -5.
v_max = 5.
v_range = v_max - v_min

hp= HernquistPotential(amp=0.5, a=1.)
def extF(x, t):
    z = abs(Jo.z(t)-x)
    return hp.zforce(0., z)

# xdensity array is created with dimensions:
# (input & output states (0 or 1), x or v (0 or 1), training data size, num. bins)
train_density = numpy.empty((2, 2, train_size, n_bins))
energy = numpy.empty((train_size, nt+1))


for j in trange(train_size, desc='Training data progress'):
    # (input & output states, phase space degrees of freedom (x & v), no. of slabs)
    train = numpy.empty((2, 2, N))
    numpy.random.seed(j)


    # compute zh based on sigma and totmass
    totmass= 1. # Sigma above
    sigma= 1.
    zh= sigma**2./totmass # twopiG = 1. in our units
    tdyn= zh/sigma
    x= numpy.arctanh(2.*numpy.random.uniform(size=N)-1)*zh*2.
    v= numpy.random.normal(size=N)*sigma
    v-= numpy.mean(v) # stabilize
    m= numpy.ones_like(x)/N#*(1.+0.1*(2.*numpy.random.uniform(size=N)-1))
    
    g= wendy.nbody(x, v, m, 0.01*tdyn, approx=True, nleap=10, t0=-2.*dt, ext_force=extF)
    

    
    train[0, 0, :]= x
    train[0, 1, :]= v
    energy[j, 0]= wendy.energy(x,v,m)
    
    for ii in arange(nt):
        tx,tv = next(g)
        energy[j, ii+1]= wendy.energy(tx,tv,m)
    
    train[1, 0, :]= tx
    train[1, 1, :]= tv
    
    

    initial_xdensity, bins = np.histogram(train[0, 0, :], bins = n_bins, range=[x_min, x_max])
    final_xdensity, bins = np.histogram(train[1, 0, :], bins = n_bins, range=[x_min, x_max])
    train_density[0, 0, j, :] =  initial_xdensity
    train_density[1, 0, j, :] =  final_xdensity

    initial_vdensity, bins = np.histogram(train[0, 1, :], bins = n_bins, range=[v_min, v_max])
    final_vdensity, bins = np.histogram(train[1, 1, :], bins = n_bins, range=[v_min, v_max])
    train_density[0, 1, j, :] =  initial_vdensity
    train_density[1, 1, j, :] =  final_vdensity
    
```

## Density Profile Datasets with Varying Perturbation Parameter & Partially Self-Gravitating Disc

### TRAIN DATA

```python
train_size = 2000
N = 1001
nt = 500
n_bins = 60
alpha= 0.3 # "live" fraction
x_min = -15.
x_max = 15.
x_range = x_max - x_min
v_min = -5.
v_max = 5.
v_range = v_max - v_min

# density array is created with dimensions:
# (input & output states (0 or 1), x or v (0 or 1), training data size, num. bins + pert param)
train_density = numpy.empty((2, 2, train_size, n_bins+1))
train_density[1, :, :, -1] = 0
energy = numpy.empty((train_size, nt+1))

count=0
for j in trange(train_size, desc='Training data progress'):

    numpy.random.seed(j+2000)

    # (input & output states, phase space degrees of freedom (x & v), no. of slabs)
    train = numpy.empty((2, 2, N))

    if j % 50 == 0:
        amp = 0.01*(count)
        hp = HernquistPotential(amp=amp, a=1.)
        count += 30
        
    train_density[0, :, j, -1] = amp

    def extF(x, t):
        z = abs(Jo.z(t)-x)
        return hp.zforce(0., z)



    # compute zh based on sigma and totmass
    totmass= 1. # Sigma above
    sigma= 1.
    zh= sigma**2./totmass # twopiG = 1. in our units
    tdyn= zh/sigma
    x= numpy.arctanh(2.*numpy.random.uniform(size=N)-1)*zh*2.
    v= numpy.random.normal(size=N)*sigma
    v-= numpy.mean(v) # stabilize
    m= numpy.ones_like(x)/N#*(1.+0.1*(2.*numpy.random.uniform(size=N)-1))
    

    # Adjust masses to only represent alpha of the mass
    m*= alpha
    # 1-alpha in the mass is then given by the external force
    sigma2= sigma**2.
    def tot_force(x,t):
        return extF(x, t)-(1.-alpha)*sigma2*numpy.tanh(0.5*x/zh)/zh

    g= wendy.nbody(x, v, m, 0.01*tdyn, approx=True, nleap=10, t0=-2.*dt, ext_force=tot_force)
    
    
    train[0, 0, :]= x
    train[0, 1, :]= v
    energy[j, 0]= wendy.energy(x,v,m)
    
    for ii in arange(nt):
        tx,tv = next(g)
        energy[j, ii+1]= wendy.energy(tx,tv,m)
        
    train[1, 0, :]= tx
    train[1, 1, :]= tv
    
    
    initial_xdensity, bins = np.histogram(train[0, 0, :], bins = n_bins, range=[x_min, x_max])
    final_xdensity, bins = np.histogram(train[1, 0, :], bins = n_bins, range=[x_min, x_max])
    train_density[0, 0, j, :-1] =  initial_xdensity
    train_density[1, 0, j, :-1] =  final_xdensity

    initial_vdensity, bins = np.histogram(train[0, 1, :], bins = n_bins, range=[v_min, v_max])
    final_vdensity, bins = np.histogram(train[1, 1, :], bins = n_bins, range=[v_min, v_max])
    train_density[0, 1, j, :-1] =  initial_vdensity
    train_density[1, 1, j, :-1] =  final_vdensity
    
```
